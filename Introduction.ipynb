```python
# Find the points where the tangent to f(t) has a slope closest to zero and depict them in the plot. Hint: approximate the tangent using the secant.
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(t):
    return np.sin(2 * np.pi * t / 5 + 0.2)

# Crear puntos para dos periodos
t = np.linspace(0, 10, 1000)  # 10 es dos periodos de 5

# Calcular los valores de la función
y = f(t)

# Calcular la derivada aproximada (pendiente de la secante)
dt = t[1] - t[0]
dy = np.diff(y) / dt

# Encontrar los puntos donde la pendiente está más cerca de cero
Umbral = 1e-3  # Umbral para considerar la pendiente cercana a cero
indices_cero = np.where(np.abs(dy) < Umbral)[0]

# Graficar
plt.figure(figsize=(12, 6))
# Graficar la función en un plot
plt.plot(t, y, label='f(t)')
# Graficar los puntos en un scatter
plt.scatter(t[indices_cero], y[indices_cero], color='red',
            label='Pendiente cercana a cero', zorder=5)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('f(t) = sin(2πt/5 + 0.2)')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir los puntos con pendiente cercana a cero
print("Puntos con pendiente cercana a cero:")
for i in indices_cero:
    print(f"t = {t[i]:.2f}, f(t) = {y[i]:.2f}")
```
